```python
### import sys
import datetime
import math

print("\n====== PYTHON ASSIGNMENT PROGRAM ======")
print("1. Twinkle Twinkle Pattern")
print("2. Show Python Version")
print("3. Current Date & Time")
print("4. Area of Circle")
print("5. Reverse First & Last Name")
print("6. Addition of Two Numbers")
print("7. Marksheet with Grades")
print("8. Even or Odd Number")

choice = int(input("\nEnter Question Number (1-8): "))

# Q1
if choice == 1:
    print("\nTwinkle, twinkle, little star,")
    print("\tHow I wonder what you are!")
    print("\t\tUp above the world so high,")
    print("\t\tLike a diamond in the sky.")
    print("Twinkle, twinkle, little star,")
    print("\tHow I wonder what you are!")

# Q2
elif choice == 2:
    print("\nPython Version:")
    print(sys.version)

# Q3
elif choice == 3:
    now = datetime.datetime.now()
    print("\nCurrent Date and Time:", now)

# Q4
elif choice == 4:
    radius = float(input("Enter radius of circle: "))
    area = math.pi * radius * radius
    print("Area of Circle:", area)

# Q5
elif choice == 5:
    fname = input("Enter First Name: ")
    lname = input("Enter Last Name: ")
    print("Reversed Name:", lname, fname)

# Q6
elif choice == 6:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    print("Addition:", num1 + num2)

# Q7 (Marksheet)
elif choice == 7:
    name = input("Enter Student Name: ")
    roll = input("Enter Roll Number: ")

    print("\nEnter Marks (out of 100 each)")
    m1 = int(input("Math: "))
    m2 = int(input("Chemistry: "))
    m3 = int(input("Islamiat: "))
    m4 = int(input("Physics: "))
    m5 = int(input("English: "))

    total = m1 + m2 + m3 + m4 + m5
    percentage = (total / 500) * 100

    print("\n----- MARKSHEET -----")
    print("Name:", name)
    print("Roll No:", roll)
    print("Obtained:", total, "/500")
    print("Percentage:", round(percentage, 2), "%")

    if percentage >= 80:
        print("Grade: A+")
    elif percentage >= 70:
        print("Grade: A")
    elif percentage >= 60:
        print("Grade: B")
    elif percentage >= 50:
        print("Grade: C")
    elif percentage >= 40:
        print("Grade: D")
    else:
        print("Grade: F")

# Q8
elif choice == 8:
    num = int(input("Enter a number: "))
    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

else:
    print("Invalid choice!")
```
