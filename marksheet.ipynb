```python
##### Simple Marksheet

# Simple Marksheet

name = input("Enter Student Name: ")
roll = input("Enter Roll Number: ")

print("\nEnter Marks (out of 100 each)\n")

math = int(input("Enter Math Marks: "))
chemistry = int(input("Enter Chemistry Marks: "))
islamiat = int(input("Enter Islamiat Marks: "))
physics = int(input("Enter Physics Marks: "))
english = int(input("Enter English Marks: "))
pak_studies = int(input("Enter Pakistan Studies Marks: "))

# Total Obtained Marks
obtained = math + chemistry + islamiat + physics + english + pak_studies

# Percentage
percentage = (obtained / 600) * 100

# -------- MARKSHEET --------
print("\n========== MARKSHEET ==========")

print("Name :", name)
print("Roll No:", roll)
print("Obtained:", obtained, "/ 600")
print("Percentage:",percentage, "%")

# Grade
if percentage >= 80:
    print("Grade: A+")
elif percentage >= 70:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 50:
    print("Grade: C")
elif percentage >= 40:
    print("Grade: D")
else:
    print("Grade : F (Fail)")
```
